```python
import pandas as pd
import matplotlib.pyplot as plt

# Caricamento del dataset (gestisce CSV o XLSX)
try:
    df_platform = pd.read_csv(file_path)
except Exception:
    df_platform = pd.read_excel(file_path)

# 1. Preparazione dei dati (Browser vs. Vittime)
df_browser = df_platform[['browser', 'victims']].copy()

# Pulisci e ordina i dati (vittime decrescenti)
df_browser['victims'] = pd.to_numeric(df_browser['victims'], errors='coerce')
df_browser = df_browser.dropna().sort_values(by='victims', ascending=False)

# 2. Creazione del Grafico a Barre Verticali
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(df_browser['browser'], df_browser['victims'], color='teal')

# Aggiungi etichette e titoli
ax.set_xlabel('Browser')
ax.set_ylabel('Numero di Vittime')
ax.set_title('Vittime di Phishing per Browser')

# Rotazione delle etichette sull'asse X per evitare sovrapposizioni
ax.set_xticks(df_browser['browser'])
ax.set_xticklabels(df_browser['browser'], rotation=0, ha='center') # Modificato: rotation=0, ha='center'

# Rimuovi la griglia e i bordi superiore e destro
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 3. Salva il grafico come PNG
output_file = 'vittime_per_browser_verticale.png'
plt.savefig(output_file, bbox_inches='tight')
plt.close()
```
