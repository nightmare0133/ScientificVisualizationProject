```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carica i dati dal file Excel
df_istruzione = pd.read_excel(file_path)

# --- Il resto del codice rimane lo stesso per la creazione del grafico ---

# 1. Definisci l'ordine logico per l'asse X (Livello di Istruzione)
istruzione_sort_order = ['Base', 'Laurea breve', 'Laurea', 'Post-laurea']

# Applica l'ordinamento categoriale per garantire che l'asse X sia ordinato correttamente
df_istruzione['Istruzione'] = pd.Categorical(
    df_istruzione['Istruzione'],
    categories=istruzione_sort_order,
    ordered=True
)
df_istruzione = df_istruzione.sort_values('Istruzione')

# 2. Creazione del Grafico a Linee (Matplotlib)
fig, ax = plt.subplots(figsize=(10, 6))

# Serie 1: Mai cliccato
ax.plot(
    df_istruzione['Istruzione'],
    df_istruzione['Mai cliccato'],
    marker='o',
    linestyle='-', 
    label='Mai cliccato'
)

# Serie 2: Mai segnalato
ax.plot(
    df_istruzione['Istruzione'],
    df_istruzione['Mai segnalato'],
    marker='o',
    linestyle='-', 
    label='Mai segnalato'
)

# 3. Aggiungi titoli e formattazione
ax.set_title('Tendenza dei Comportamenti di Phishing per Livello di Istruzione')
ax.set_xlabel('Livello di Istruzione')
ax.set_ylabel('Proporzione')
ax.grid(False) # Rimosse le righe della griglia
ax.legend(title='Comportamento')
ax.set_ylim(0, df_istruzione[['Mai cliccato', 'Mai segnalato']].max().max() * 1.1)

# Rimuovi i bordi superiore e destro
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 4. Salva il grafico come PNG
output_file = 'grafico_tendenza_istruzione.png'
plt.savefig(output_file, bbox_inches='tight')
plt.close()
```
